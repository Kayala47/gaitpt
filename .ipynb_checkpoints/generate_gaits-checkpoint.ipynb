```python
%%capture
# Suppress animation plot

n = 32
t = np.linspace(0, 2, n)

foot_swinging = False
foot_stride = 1.5
foot_halfstride = foot_stride / 2

hip_x = np.linspace(1, 9, n)
foot_x = hip_x[0] + foot_halfstride

xlim = [0, 10]
ylim = [-0.5, 1.5]

fig, ax = plt.subplots()
(leg,) = ax.plot([], [], marker="o", linewidth=5)
ax.plot(xlim, [0, 0], "--", linewidth=5)


def anim_init():
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    #     ax.set_yticks([])
    #     ax.set_xticks([])
    return (leg,)


def anim_update(frame):
    global foot_x, foot_swinging

    leg.set_data([hip_x[frame], foot_x], [1, 0.2 if foot_swinging else 0])

    # Leg swinging state machine
    if not foot_swinging and foot_x < hip_x[frame] - foot_halfstride:
        foot_swinging = True
    elif not foot_swinging:
        pass
    elif foot_swinging and foot_x > hip_x[frame] + foot_halfstride:
        foot_swinging = False
    else:
        foot_x += 8 / (n / 3)

    return (leg,)


anim = FuncAnimation(fig, anim_update, frames=n, init_func=anim_init, blit=True)
```
